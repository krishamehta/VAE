#### Model 'CVIT' distribution as a deterministic function of two simple distribution,
#### namely, the canonical Gaussian and the multi-nomial distribution.

```python
def cvit(z12, z3):
    os = 10
    ost = np.pi/2
    
    x = np.zeros((len(z3),2))
    
    for i in range(len(z3)):
        z1 = z12[i,0]
        z2 = z12[i,1]
        
        idx = z3[i]
        
        if idx==0:
            # C in CVIT
            x2 = (os+z1)*np.cos(z2+ost)
            x1 = -(1.5*os+z1)*np.abs(np.sin(z2+ost))
        elif idx==1:
            # V in CVIT
            x1 = 20+(os+z1)*np.cos(z2+ost)
            x2 = 12-0.6*(4*os+z1)*np.abs(np.sin(z2+ost))
        elif idx==2:
            # I
            x2 = (10+z1)*np.cos(z2+ost)
            x1 = 40+(1+z1)*(np.sin(z2+ost))
        elif idx==3:
            # T_h
            x1 = 60+(10+z1)*np.cos(z2+ost)
            x2 = 10+(1+z1)*(np.sin(z2+ost))
        elif idx==4:
            # T_v
            x2 = (10+z1)*np.cos(z2+ost)
            x1 = 60+(1+z1)*(np.sin(z2+ost))

        x[i,0] = x1
        x[i,1] = x2
        
    return x
```

```python
numSamples = 2000
z12 = np.random.randn(numSamples,2)
z3 = np.random.randint(5,size=(numSamples,1))

x = cvit(z12,z3)

x1 = x[:,0];
x2 = x[:,1];

plt.scatter(x1,x2)
plt.hold()
plt.grid(True)
plt.show()
```
